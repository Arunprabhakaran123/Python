```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path="D:/Train/train.csv"
```

```python
df = pd.read_csv('D:/Train/train.csv')
```

```python
df
```

```python
#1 histogram for numeric columns

numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

sns.set(style="whitegrid")

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

```

```python
#2 count plot caterogical columns

sns.set(style="whitegrid")

categorical_cols = ['Sex', 'Embarked', 'Pclass', 'Survived']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
#3 box plot survival vs numerical columns

sns.set(style="whitegrid")
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
for col in numeric_cols:
    plt.figure(figsize=(7, 5))
    sns.boxplot(x='Survived', y=col, data=df, palette='Set3')
    plt.title(f'{col} Distribution by Survival')
    plt.xlabel('Survived (0 = No, 1 = Yes)')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
# 4 scatter plot pairwish survival visualization


numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df_clean = df[numeric_cols + ['Survived']].dropna()
sns.pairplot(df_clean, hue='Survived', palette='Set1', diag_kind='kde')
plt.suptitle('Pairwise Scatter Plot by Survival', y=1.02)
plt.show()
```

```python
#5 survival rate by category 

categorical_cols = ['Sex', 'Pclass', 'Embarked']
for col in categorical_cols:
    survival_rate = df.groupby(col)['Survived'].mean().sort_values(ascending=False)

    plt.figure(figsize=(6, 4))
    sns.barplot(x=survival_rate.index, y=survival_rate.values, palette='viridis')
    plt.title(f'Survival Rate by {col}')
    plt.ylabel('Survival Rate')
    plt.xlabel(col)
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()
```

```python
#6 correlation heat map

numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
df_numeric = df[numeric_cols].dropna()
corr_matrix = df_numeric.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```
